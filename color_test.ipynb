# Setup

```python
import numpy as np
from PIL import Image, ImageOps

from MyHybridImages import myHybridImages

%matplotlib inline
from matplotlib import pyplot as plt
```

# Get Image Paths

```python
lowImage = np.asarray(Image.open('data/dog.bmp'))
highImage = np.asarray(Image.open('data/cat.bmp'))
```

# Create Hybrid Image

```python
hybridImage = myHybridImages(lowImage=lowImage, lowSigma=2.5, highImage=highImage, highSigma=7)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(lowImage)
plt.title('lowImage')
plt.yticks([])

plt.subplot(1, 3, 2)
plt.imshow(highImage)
plt.title('highImage')
plt.yticks([])

plt.subplot(1, 3, 3)
plt.imshow(hybridImage)
plt.title('hybridImage')
plt.yticks([])
```

```python
im = Image.fromarray(hybridImage.astype(np.uint8), 'RGB')
```

# Cascading Hybrid Image

```python
def resize_image (img, baseheight):
    hpercent = (baseheight / float(img.size[1]))
    wsize = int((float(img.size[0]) * float(hpercent)))
    return img.resize((wsize, baseheight), Image.Resampling.LANCZOS)
```

```python
im_1 = resize_image(im, 300)
im_2 = resize_image(im, 150)
im_3 = resize_image(im, 80)
im_4 = resize_image(im, 40)
```

```python
def display_image_in_actual_size(im_data):

    dpi = 80
    height, width = im_data.size

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    # Display the image.
    plt.figure(figsize=figsize)
    plt.imshow(im_data)

    plt.show()
```

```python
display_image_in_actual_size(im_1)
display_image_in_actual_size(im_2)
display_image_in_actual_size(im_3)
display_image_in_actual_size(im_4)
```
