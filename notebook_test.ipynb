```python
# im = Image.fromarray(hybridImage.astype(np.uint8))
# im.save('output.jpg')
im = Image.fromarray(hybridImage[:,:,0])
#im.convert('RGB').save('output.jpg')
```

```python
def resize_image (img, baseheight):
    hpercent = (baseheight / float(img.size[1]))
    wsize = int((float(img.size[0]) * float(hpercent)))
    return img.resize((wsize, baseheight), Image.LANCZOS)
```

```python
im_1 = resize_image(im, 300)
im_2 = resize_image(im, 150)
im_3 = resize_image(im, 80)
im_4 = resize_image(im, 40)
```

```python
def display_image_in_actual_size(im_data):

    dpi = 80
    height, width = im_data.size

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data)

    plt.show()
```

```python
display_image_in_actual_size(im_1)
display_image_in_actual_size(im_2)
display_image_in_actual_size(im_3)
display_image_in_actual_size(im_4)
```
